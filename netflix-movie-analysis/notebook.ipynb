```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
print(netflix_df.head()) # netflic dataset
# Filter the 90s movies in Netflix
fmov90s = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] < 2000)]
print("Movies of 1990s are: ")
print(fmov90s) #90s movie dataframe


# Frequent movie duration
plt.hist(fmov90s["duration"])
plt.xlabel('Duration')
plt.ylabel('no. of movies')
plt.title('Distribution of Movie Durations on Netflix (90s only)')
plt.show()

duration = 100
#subsetting action movies
action_movies = fmov90s[fmov90s['genre']== "Action"]
#count the number of action movies from the 90s
short_movie_count = 0
for label, row in action_movies.iterrows():
    if row["duration"]< 90 :
        short_movie_count =short_movie_count+ 1
    else:
        short_movie_count= short_movie_count
print("Number of short action movies in 90s are: " + str(short_movie_count))
```
